**CHECKING FOR MISSING VALUES AND HANDLING IT

```python
#lets check if there is any missing values available in the datatset

df.isnull().sum()

#so we can see below that 4 of the columns have missing values as well
#column children has 4  missing values 
#column country has 488 missing values
#column agent has 16340 missing values
#column company has 112593 missing values
```

```python
#we saw above that agent has 16340 missing values and company has 112593 missing values in the dataset

#idiolly we should remove the missing vales but as the missing values are in huge amount so its not a good idea 
#to drop all the rows which has missing values rather we can remove both the columns itself as these are not very important columns for our analysis
#For other columns the amount of missing values are less comparatively so we can remove thier respective rows but not in these 2 columns.


df.drop(['agent','company'],axis=1,inplace=True)   #removing both the columns
         
df.dropna(inplace=True)  #removing other missing records(rows)
```

```python
#Now we can see all the missing values has been removed
df.isnull().sum()
```

**CHECKING THE DATATYPES OF ALL THE COLUMNS

```python
#Now lets check the data types of all the columns as having the correct data types is also very important

df.info()

```

```python
#We can see above that the data type of the column named reservation_status_date is object
#We know that any date should have datetime datatype
#So lets change its datatype from object to datetime

df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])

```
